## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah ada peran dari cuaca terhadap jumlah pengguna sepeda?

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='weathersit',
    y='cnt',
    data=bike_day_df)

plt.title('Jumlah Pengguna Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca. Ket: 1 = Clear, 2 = Cloudly, 3 = Rain')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

### Pertanyaan 2: Bagaimana tren penggunaan sepeda dalam setahun, Diantara tahun 2011 dan 2012 manakah yang lebih unggul?

```python
bike_day_df['dteday'] = pd.to_datetime(bike_day_df['dteday'])

bike_day_df['yr'] = bike_day_df['dteday'].dt.year

plt.figure(figsize=(10, 6))
plt.plot(bike_day_df['dteday'], bike_day_df['cnt'])
plt.title('Tren Peminjaman Sepeda pada Tahun 2011 dan 2012')
plt.xlabel('Tanggal')
plt.ylabel('Total Peminjaman Sepeda')
plt.show()
```

## Conclusion
